# Задание 4.2 Перевозки (статистика)

Возьмите данные о перевозках между посёлками.

Проведите небольшое статистическое исследование,  для визуализации результатов постройте не менее трех диаграмм и графиков (например, круговую диаграмму, показывающую суммарный расход бензина по датам по всем направлениям).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('task19.csv', encoding='cp1251', delimiter=';')
data
```

Построим диаграмму, показывающую суммарный расход бензина по датам по всем направлениям

```python
petrol = data.groupby('Дата')['Расход бензина'].sum()
```

```python
petrol.plot.pie(autopct='%.2f%%', figsize=(10, 10))
plt.title('Суммарный расход бензина по датам по всем направлениям')
```

Построим диаграмму зависимости расхода бензина от расстояния:

```python
plt.figure(figsize=(10, 10))
plt.scatter(x=data['Расстояние'], y=data['Расход бензина'])
plt.title('Зависимость расхода бензина от расстояния')
plt.xlabel('Расстояние')
plt.ylabel('Расход бензина')
```

Теперь визиализируем диаграмму средней массы грузы по пунктам отправления:

```python
data.groupby('Пункт отправления')['Масса груза'].mean().plot.bar()
plt.title('Средняя масса груза по пунктам отправления')
```
